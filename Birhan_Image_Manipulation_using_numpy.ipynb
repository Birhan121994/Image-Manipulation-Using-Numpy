### Menu3: Convert RGB(colored) image to Grey scale image

```python
import cv2
import numpy as np
from PIL import Image
img = np.array(cv2.imread("superhero.png"))
negativeImg = 255 - img
img_array = img.copy()

for i in range(len(img_array)):
    for j in range(len(img_array[i])):
        blue = img_array[i,j,0]
        green = img_array[i,j,1]
        red = img_array[i,j,2]
        
        grey_scale = (red + green + blue / 3)
        img_array[i,j] = grey_scale

grey_scale_img = Image.fromarray(img_array)
# grey_scale_img.save("grey_scale_image.jpeg")
grey_scale_image = cv2.imread('grey_scale_image.jpeg')

frame1 = plt.figure(figsize=(15,4));

axis_1 = frame1.add_subplot(1,4,1);
axis_1.imshow(img);
plt.title("The Original Image");
plt.xticks([]);
plt.yticks([]);

axis_2 = frame1.add_subplot(1,4,2);
axis_2.imshow(negativeImg);
plt.title("Negative of the original image")
plt.xticks([]);
plt.yticks([]);

axis_3 = frame1.add_subplot(1,4,3);
axis_3.imshow(grey_scale_image);
plt.title("Grey Scale Image")
plt.xticks([])
plt.yticks([])
# cv2.imshow("The Original Image",img)
# cv2.imshow("Negative of the original image",negativeImg)
# cv2.imshow("Grey Scale Image",grey_scale_image)
# cv2.waitKey(0)
```

### Menu4: Create a new image by adding images together

```python
import cv2
import numpy as np
from PIL import Image

img1 = np.array(cv2.imread("Image1.jpg"))
img2 = cv2.imread("Image2.jpg")
res_img = cv2.resize(img2,img1.shape[1::-1])
New_Image = (img1+ res_img).astype(np.uint8)

frame3 = plt.figure(figsize=(15,4));

axis_num_1 = frame3.add_subplot(1,4,1)
axis_num_1.imshow(img1);
plt.title("The First Image");
plt.xticks([]);
plt.yticks([]);

axis_num_1 = frame3.add_subplot(1,4,2)
axis_num_1.imshow(res_img);
plt.title("The Second Image");
plt.xticks([]);
plt.yticks([]);

axis_num_1 = frame3.add_subplot(1,4,3)
axis_num_1.imshow(New_Image);
plt.title("The New Blended Image");
plt.xticks([]);
plt.yticks([]);




# cv2.imshow("The First Image",img1)
# cv2.imshow("The Second Image",img2)
# cv2.imshow("The New Image",New_Image)
# cv2.waitKey(0)
```
